```python
import os
import shutil
import pandas as pd

# Carregar os dados
train_data = pd.read_csv(r"C:\Users\lucas.costa\Documents\Human Action Recognition\Training_set.csv")

# Verificar os rótulos e suas contagens
print("Contagem de rótulos:")
print(train_data['label'].value_counts())

# Caminho da pasta onde as imagens estão armazenadas
images_folder = r"C:\Users\lucas.costa\Documents\Human Action Recognition\train"

# Verificar se a pasta de imagens existe
if not os.path.exists(images_folder):
    print(f"Erro: A pasta de imagens '{images_folder}' não existe.")
    exit()

# Caminho da pasta de destino (onde as imagens selecionadas serão salvas)
destination_folder = r"C:\Users\lucas.costa\Documents\Human Action Recognition\selected_images"

# Criar a pasta de destino se não existir
if not os.path.exists(destination_folder):
    os.makedirs(destination_folder)
    print(f"Pasta de destino '{destination_folder}' criada.")

# Função para extrair e salvar imagens
def extract_and_save_images(data, label, num_images, images_folder, destination_folder):
    # Filtrar as imagens com o rótulo desejado
    filtered_images = data[data['label'] == label]
    
    # Verificar se há imagens com o rótulo especificado
    if filtered_images.empty:
        print(f"Nenhuma imagem encontrada para o rótulo '{label}'.")
        return
    
    # Selecionar as primeiras 'num_images' imagens
    selected_images = filtered_images.head(num_images)
    
    # Copiar as imagens para a pasta de destino
    for idx, row in selected_images.iterrows():
        image_name = row['filename']  # Use 'filename' como nome da coluna
        image_path = os.path.join(images_folder, image_name)
        destination_path = os.path.join(destination_folder, f"{label}_{image_name}")
        
        # Verificar se a imagem existe antes de copiar
        if os.path.exists(image_path):
            shutil.copy(image_path, destination_path)
            print(f"Copied {image_name} to {destination_path}")
        else:
            print(f"Erro: Imagem {image_name} não encontrada em {images_folder}.")

# Extrair e salvar 15 imagens do rótulo 'drinking'
extract_and_save_images(train_data, 'drinking', 15, images_folder, destination_folder)

# Extrair e salvar 15 imagens do rótulo 'laughing'
extract_and_save_images(train_data, 'laughing', 15, images_folder, destination_folder)

print("Processo concluído!")
```
